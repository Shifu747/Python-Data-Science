```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0,5,11)
y = x**2
```

# Functional Method

```python
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

#subplot (no. of rows, no. of cloumns, plot no.)
```

## OO

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #[left bottom width height]

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
#axes3 = fig.add_axes([0.4,0.15,0.4,0.3])


axes1.plot(x,y)

axes1.set_title('LARGER PLOT')

axes2.plot(y,x)

axes2.set_title('SMALLER PLOT')
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label= 'X Cubed')
ax.plot(x,x**.5, label= 'X SQRT')

ax.legend(loc=0)
```

## Part 3

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',lw=1,alpha=.5, ls='-',marker='o',markersize=20,
       markerfacecolor='red',markeredgewidth=6,markeredgecolor='green')

#color takes hex code 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='black',lw=1,alpha=1, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
